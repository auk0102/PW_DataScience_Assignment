###1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random phenomenon. It maps outcomes of a probabilistic experiment to real numbers.
- Discrete random variable: Takes countable values (e.g., number of heads in 10 coin tosses).
- Continuous random variable: Takes uncountably infinite values (e.g., weight, height, temperature).

###2. What are the types of random variables?
- Discrete random variables: Finite or countably infinite outcomes (e.g., rolling a die).
- Continuous random variables: Infinite and uncountable outcomes (e.g., time, distance).

###3. What is the difference between discrete and continuous distributions?

- Discrete distributions are defined over countable outcomes (e.g., binomial, Poisson).


- Continuous distributions are defined over intervals of real numbers (e.g., normal, exponential).


- Discrete use PMF (probability mass function); continuous use PDF (probability density function).

###4. What are probability distribution functions (PDF)?

A PDF describes the likelihood of a continuous random variable taking on a specific value.
- For continuous variables, the probability of an exact value is 0; we calculate probabilities over intervals.
- The area under the PDF curve equals 1.

###5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- CDF: Gives the probability that a random variable is less than or equal to a value

- PDF: Describes the relative likelihood of a specific outcome (for continuous variables).

CDF is the integral of PDF.

###6. What is a discrete uniform distribution?

A distribution in which all outcomes have equal probability.

###8. What is the binomial distribution, and how is it used in probability?

- It models the number of successes in a fixed number of repeated independent trials, where each trial has the same chance of success. It's used in situations like quality control or survey responses.

###9. What is the Poisson distribution and where is it applied?
This distribution is used to model the number of times an event occurs within a fixed period or space, especially when the events happen independently and at a constant rate. Itâ€™s often used in fields like telecommunications or traffic flow.

- Models the number of events in a fixed interval of time or space.

- Events occur independently at a constant average rate

- Used in queuing theory, telecom, natural events (earthquakes, calls).

###10. What is a continuous uniform distribution?

It describes a situation where all values within a certain range are equally likely to occur. It's like spinning a fair wheel where every spot is just as likely to be selected.
- All intervals of the same length in the distribution's range are equally probable.


###11. What are the characteristics of a normal distribution?
- Bell-shaped, symmetric around the mean.

- Mean = median = mode.

- Follows the empirical rule (68%-95%-99.7%).

It is a symmetrical, bell-shaped curve where most values cluster around the middle, and fewer values appear as you move away from the center. It represents many natural phenomena, such as height or test scores.

###13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- States that the sampling distribution of the sample mean approaches normality as sample size increases, regardless of the original distribution.

Important because it allows for normal approximation even for non-normal populations.

###14. How does the Central Limit Theorem relate to the normal distribution?
- It shows that no matter what the original data looks like, the averages of large samples will be normally distributed. This connection makes the normal distribution a powerful tool in analysis.

###15. What is the application of Z statistics in hypothesis testing?
- Z-statistics measure how many standard deviations an observation is from the mean.

Used to test hypotheses about population means when variance is known and sample size is large.

###17. What are point estimates and interval estimates in statistics?
- Point estimate: A single value used to estimate a population parameter (e.g., sample mean).

- Interval estimate: A range (e.g., confidence interval) likely to contain the population parameter.

###18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals help us understand the reliability of an estimate. They show a range of likely values for a population parameter and give a sense of how sure we are about our estimate. Provide a range of plausible values for a parameter.

E.g., a 95% CI means that if we repeated the experiment many times, 95% of the intervals would contain the true parameter.

###19. What is the relationship between a Z-score and a confidence interval?
- Z-scores are used to create confidence intervals. They help set the boundaries that define how far out we need to go from the average to capture a certain level of confidence

###20. How are Z-scores used to compare different distributions?
